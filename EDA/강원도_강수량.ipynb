# **1993~2019 강수량**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv('/content/gangwon_rain.csv',sep=',',encoding='cp949')

data.tail()
```

# **연도별 월 평균 강수량**

```python
years_by_mean=data.groupby('연도별')[['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']].agg(['mean'])
```

```python
years_by_mean.head()
```

```python
a1=years_by_mean

fig,axes = plt.subplots(4,1,figsize=(6,15));
axes[0].plot(a1['1월']);axes[0].set_title('Jan')
axes[1].plot(a1['2월']);axes[1].set_title('Feb')
axes[2].plot(a1['3월']);axes[2].set_title('Mar')
axes[3].plot(a1['4월']);axes[3].set_title('Apr')


fig.tight_layout()
```

```python
fig,axes = plt.subplots(4,1,figsize=(6,15));
axes[0].plot(a1['5월']);axes[0].set_title('May')
axes[1].plot(a1['6월']);axes[1].set_title('Jun')
axes[2].plot(a1['7월']);axes[2].set_title('Jul')
axes[3].plot(a1['8월']);axes[3].set_title('Aug')

fig.tight_layout()
```

```python
fig,axes = plt.subplots(4,1,figsize=(6,15));
axes[0].plot(a1['9월']);axes[0].set_title('Sep')
axes[1].plot(a1['10월']);axes[1].set_title('Oct')
axes[2].plot(a1['11월']);axes[2].set_title('Nav')
axes[3].plot(a1['12월']);axes[3].set_title('Dec')

fig.tight_layout()
```
