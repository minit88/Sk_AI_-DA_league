```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/OBS_ASOS_MNH_20221022010713.csv',encoding='cp949')
data=data.drop(columns = ['지점','지점명'])

data.head()
```

```python
labels=[]

for i in range(len(data['일시'])):
  if i%10==0 :
    labels.append(data['일시'][i])
print(labels)
```

# **강릉 월별 평균기온(2015~2022)**

```python
plt.figure(figsize=(10,8));plt.plot(data['일시'],data['평균기온(°C)'])
plt.grid(True)
plt.title('Year_by_month_Avg_temperature')
plt.xticks(labels,labels)
```

# **원주 월별 평균기온(2015~2022)**

```python
data1= pd.read_csv('/content/OBS_ASOS_MNH_20221022010810.csv',encoding='cp949')
data1.tail()
```
