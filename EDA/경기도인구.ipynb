```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import os
import natsort

from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/skt 데이터/data/경기도인구수/'
file_list = os.listdir(path)
file_list = [file for file in file_list if file.endswith('.csv')]
file_list
```

```python
for i in file_list:
    data = pd.read_csv(path + i,encoding='UTF8')
  
data.head()
```

```python
df = pd.DataFrame()
data = data.drop([0,1])
df['2018'] = data['2018'].astype(int) + data['2018.1'].astype(int)
df['2019'] = data['2019'].astype(int) + data['2019.1'].astype(int)
df['2020'] = data['2020'].astype(int) + data['2020.1'].astype(int)
df['2021'] = data['2021'].astype(int) + data['2021.1'].astype(int)

df = df.set_index(keys=data['시군구별(1)'])
df
```
