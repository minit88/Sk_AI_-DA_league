```python
path = '/content/drive/MyDrive/skt 데이터/data/웨더아이/봄꽃/'

file_list = os.listdir(path)
file_list = [file for file in file_list if file.endswith('.csv')]
file_list
```

```python
yellow = pd.DataFrame()
pink = pd.DataFrame()

for i in file_list:
  data = pd.read_csv(path + i, encoding='cp949')
  yellow = pd.concat([yellow, data['개나리']], axis=1)
  pink = pd.concat([pink, data['진달래']], axis=1)

yellow = pd.concat([yellow, data['지역명']], axis = 1)
yellow = yellow.set_index(keys='지역명')

pink = pd.concat([pink, data['지역명']], axis = 1)
pink = pink.set_index(keys='지역명')

yellow.columns=["2018", "2019", "2020", "2021", '2022']
pink.columns=["2018", "2019", "2020", "2021", '2022']
```

```python
path = '/content/drive/MyDrive/skt 데이터/data/웨더아이/벚꽃/'

file_list = os.listdir(path)
file_list = [file for file in file_list if file.endswith('.csv')]
file_list
```

```python
blossom = pd.DataFrame()

for i in file_list:
  data = pd.read_csv(path + i, encoding='cp949')
  blossom = pd.concat([blossom, data['벚꽃']], axis=1)

blossom = pd.concat([blossom, data['주요도시']], axis = 1)
blossom.rename(columns = {'주요도시':'지역명'}, inplace = True)

blossom = blossom.set_index(keys='지역명')
blossom.columns=["2018", "2019", "2020", "2021", '2022']
```

**전국 벚꽃 개화시기 (5개년)**

```python
blossom
```

**충청남도 (대전) 개나리, 진달래, 벚꽃 개화시기 (5개년)**

```python
대전 = pd.concat([yellow.filter(like='대전',axis=0), pink.filter(like='대전',axis=0), blossom.filter(like='대전',axis=0)])
대전 = 대전.set_axis(['개나리', '진달래','벚꽃'], axis=0)
대전
```
